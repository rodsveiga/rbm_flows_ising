## Magnetization near $T_c$ as a function of $L$

```python
def plot_thermodynamics(L, 
                        spin_MC, 
                        Tc_scaled= False):
    
    T_c = 2 / ( np.log(1 + np.sqrt(2)) )
    N_spins = L*L
    
    if Tc_scaled:
        spin_MC['temp'] = spin_MC['temp'] / T_c
        
    expec_val_per_spin_MAGN = abs(spin_MC['magn']).groupby(spin_MC['temp']).mean() / N_spins
                   
    temp_array = np.array(expec_val_per_spin_MAGN.index, dtype = np.float64)
        
    expec_val_MAGN  = N_spins*expec_val_per_spin_MAGN 
        
       
    plt.plot(expec_val_per_spin_MAGN, 'o-', label= 'L = %d' % L)
    
    if Tc_scaled:
        plt.axvline(x= 1.0, color='k', linestyle='--')
        plt.xlabel('$T / T_c$', fontsize=14)
    else:
        plt.axvline(x= T_c, color='k', linestyle='--') 
        plt.xlabel('$T$', fontsize=14)
        
    plt.ylabel('$<M>/N$', fontsize=14)
```

```python
plot_thermodynamics(L= 10,
                    spin_MC= ising_L010, 
                    Tc_scaled= False)
```
