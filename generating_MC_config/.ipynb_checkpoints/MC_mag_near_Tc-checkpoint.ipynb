## Magnetization near $T_c$ as a function of $L$

```python
def mag_computation(L,
                    spin_MC,
                    Tc_scaled= False):
    
    T_c = 2 / ( np.log(1 + np.sqrt(2)) )
    N_spins = L*L
    
    if Tc_scaled:
        spin_MC['temp_'] = spin_MC['temp'] / T_c
        
    else:
        spin_MC['temp_'] = spin_MC['temp'] 
               
    expec_val_per_spin_MAGN = abs(spin_MC['magn']).groupby(spin_MC['temp_']).mean() / N_spins
    
    return expec_val_per_spin_MAGN
```

```python
publication = False
save = False

    
if publication:
    plt.figure(figsize=(3.375, 0.678*3.375))
    plt.xticks(fontsize= 6)
    plt.yticks(fontsize= 6)
    markersize = 1.5
    linewidth= 0.4
    fontsize_l = 7.0
    fontsize_t = 8.0
    fontsize_lg= 5.0
    linewidth_v = 0.5
    path_fig = '../figures/PUB_MAG_MC_near_Tc.pdf'
    
else:
    markersize= 5.0
    linewidth= 0.8
    fontsize_l= 12.0
    fontsize_t= 12.0
    fontsize_lg= 10.0
    linewidth_v = 1.0
    path_fig = '../figures/MAG_MC_near_Tc.pdf'
        

L_list = [10, 36, 48, 100]

ising_list = [ising_L010, ising_L036, ising_L048, ising_L100 ]


for L, ising in zip(L_list, ising_list):
    
    magn = mag_computation(L= L, 
                           spin_MC= ising,
                           Tc_scaled= True)
    
    plt.plot(magn, 
             '-o',
             label= 'L = %d' % L,
             markersize= markersize,
             linewidth= linewidth)

    
plt.axvline(x= 1.0, linestyle= ':', linewidth= linewidth_v, color= 'black')

plt.xlabel('$T / T_c$', fontsize= fontsize_l)
plt.ylabel('$<M>/N$', fontsize= fontsize_l)
plt.legend(fontsize= fontsize_lg)

if save:
    plt.savefig(path_fig , bbox_inches='tight')
plt.show()
```
